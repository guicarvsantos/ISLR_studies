- A regresão linear assume que existe uma relação aproximadamente linear entre X e Y.
<br>
- Expressando matematicamente:

<center>$Y \approx \beta_{0} + \beta_{1}*X $<center>
   
onde $\beta_{0}$ e $\beta_{1}$ sao conhecidos como coenficientes ou parametros

## Estimando os Coeficientes

- O objetivo é encontrar $\beta_{0}$ e $\beta_{1}$ tais que o modelo melhor aproxime os dados
- Contudo existem uma série de formas que podemos avaliar a "proximidade" com os pontos
- A forma mais comum de fazer isso envolve minimizar o critério dos minimos quadrados
- Outros critérios do Capitulo 6.



<center>No qual $\hat y_{i} = \beta_{0} + \beta_{1}*x_{i}$ é a prediçao do i-ésimo valor de x e $e_{i} = y_{i} - \hat y_{i}$</center>

Podemos definir como soma dos quadrados residuais (RSS):
<center>$RSS = e^2_{1} + e^2_{2} + ... + e^2_{n} =\sum_{i=1}^{n} (\hat y_{i} - \beta_{0} + \beta_{1}*x_{i})^2$</center>
Dessa forma podemos escolher o $\beta_{0}$ e $\beta_{1}$ que minimizam o RSS:
<center>$ \hat\beta_{1} = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - x)^2} $ </center>
<center>$ \hat\beta_{0} = \bar{y} - \hat\beta_{1}*\bar{x} $ </center>
